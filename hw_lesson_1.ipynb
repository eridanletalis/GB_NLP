3. Заменим сокращения с апострофами (пример: ain't, can't) на пробел, используя apostrophe_dict. Для этого необходимо сделать функцию: для каждого слова в тексте проверить (for word in text.split()), если слово есть в словаре apostrophe_dict в качестве ключа (сокращенного слова), то заменить ключ на значение (полную версию слова).

```python
for i in range(len(combine_df)):
    text = combine_df.loc[i, 'tweet']
    search_result = re.findall(r"[a-z]+'[a-z]+", text)
    new_text = []
    if search_result:
        splitted_text = re.split(r" ", text)
        for word in splitted_text:
            if re.match(r"[a-z]+'[a-z]+", word):
                if word in apostrophe_dict:
                    new_text.append(apostrophe_dict[word])
                else:
                    new_text.append(" ")
              
            else:
                new_text.append(word)

        combine_df.loc[i, 'tweet'] = new_text
```

```python
def preprocessing_step_3(df):
    for i in range(len(df)):
        text = df.loc[i, 'tweet']
        search_result = re.findall(r"[a-z]+'[a-z]+", text)
        new_text = []
        if search_result:
            splitted_text = re.split(r" ", text)
            for word in splitted_text:
                if word in ('', ' ', '  '): # Убираем лишние пробелы
                    continue
                if re.match(r"[a-z]+'[a-z]+", word):
                    if word in apostrophe_dict:
                        new_text.append(apostrophe_dict[word].lower()) # Убираем большие буквы, которые были в словаре
                    else:
                        new_text.append(" ")

                else:
                    new_text.append(word)

            df.loc[i, 'tweet'] = " ".join(new_text)
```

4. Заменим сокращения на их полные формы, используя short_word_dict. Для этого воспользуемся функцией, используемой в предыдущем пункте.

5. Заменим эмотиконы (пример: ":)" = "happy") на пробелы, используя emoticon_dict. Для этого воспользуемся функцией, используемой в предыдущем пункте.

```python
def preprocessing_steps_4_5(df):
    for i in range(len(df)):
        text = df.loc[i, 'tweet']
        new_text = []
        splitted_text = re.split(r" ", text)
        for word in splitted_text:
            if word in short_word_dict:
                new_text.append(short_word_dict[word])

            elif word in emoticon_dict:
                 new_text.append(emoticon_dict[word])

            elif re.match(r"[a-z]+\W+[a-z]+", word):
                strange = re.split(r"\W+", word)
                for w1 in strange:
                    if w1 in short_word_dict:
                         new_text.append(short_word_dict[w1])
                    else:
                        if w1 not in ('', ' ', '  '):
                             new_text.append(w1)
            else:
                if word not in ('', ' ', '  '):
                    new_text.append(word)

        df.loc[i, 'tweet'] = " ".join(new_text)
    
```

```python
def preprocessing_steps_9(df):
    for i in range(len(df)):
        text = df.loc[i, 'tweet']
        df.loc[i, 'tweet'] = ' '.join([w for w in text.split() if len(w)>1])
```

11. Удалим стоп-слова из токенов, используя nltk.corpus.stopwords. Создадим столбец 'tweet_token_filtered' без стоп-слов.

```python
from nltk.corpus import stopwords
# nltk.download('stopwords')
# @lru_cache(maxsize=128)
def wo_stop_worlds(tokens, stop_worlds=stopwords.words('english')):
    return [t for t in tokens if not t in stop_worlds]
```

12. Применим стемминг к токенам с помощью nltk.stem.PorterStemmer. Создадим столбец 'tweet_stemmed' после применения стемминга.

```python
# from nltk.stem.snowball import SnowballStemmer
# snowball = SnowballStemmer('english')
```

```python
# @lru_cache(maxsize=128)
def stemming_processor(tokens, stemmer=PorterStemmer()):
    return [stemmer.stem(t) for t in tokens]
```

13. Применим лемматизацию к токенам с помощью nltk.stem.wordnet.WordNetLemmatizer. Создадим столбец 'tweet_lemmatized' после применения лемматизации.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
# @lru_cache(maxsize=128)
def lemmatizer(tokens, lemmatizer=WordNetLemmatizer(), pos=wordnet.VERB):
    return [lemmatizer.lemmatize(word, pos) for word in tokens]
```
